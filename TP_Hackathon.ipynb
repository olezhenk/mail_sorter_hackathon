Хакатон выполняется командами от 3 до 5 человек. В то же время следует помнить, что чем больше в команде людей, тем выше предъявляются требования в целом к команде, а также к выполняемому ею проекту. Команды разрешается формировать из студентов разных групп и разных потоков.

При оценивании результатов выполнения проекта командой в обязательном порядке проверяется и учитывается в оценке вклад **каждого участника** в общую работу команды. Вследствие этого крайне важно ответственно подойти к выбору сокомандников, распределению обязанностей внутри команды, а также пресечению наличия "мертвых душ" в команде.

При необходимости команда большинством голосов имеет право исключить человека из команды **до загрузки работ** в систему [Anytask](https://anytask.org/), сообщив об этом преподавателю, в случае если такой человек никак не проявлялся в ходе выполнения проекта и являлся как раз таки той самой "мертвой душой". Помните про эту возможность, поскольку наличие "мертвых душ" может негативно повлиять на качество и оценку для всего проекта и, как следствие, в целом для всей команды!

## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке). Кроме того, в исключительных случаях, при демонстрации участником на защите превосходящих ожидания уровней понимания, участия и вовлечения в рамках проекта, комиссия может своим решением повысить индивидуальную оценку для этого участника на 0,5 или 1 балл относительно общей оценки за проект (первой оценки).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую опубликовано задание хакатона, например, `np.random.seed(29052026)`.

Обратите, пожалуйста, внимание на то, что бОльшая часть комиссий осуществляет приём проектов в очной форме, однако могут присутствовать комиссии, которые будут работать дистанционно с использованием средств видеосвязи. В случае попадания команды путем случайного распределения в комиссию, работающую в очной форме, *не допускается* какое-либо изменение формата проведения для команды на онлайн-режим, равно как и любые другие перемещения команд между комиссиями (смены комиссий), для сохранения честности и непредвзятости случайного распределения.

Явка студента на защиту проекта является обязательной; вклад студента без его присутствия не оценивается и автоматически является равным нулю. Для дистанционных комиссий дополнительным требованием является обязательное наличие включенной камеры в течение всего времени нахождения в конференции

## Описание задания

### Бизнес-контекст

Отдел IT-поддержки компании ежедневно получает сотни обращений по электронной почте. Сотрудники тратят значительное время на ручную сортировку входящих писем: критические инциденты могут теряться среди рутинных запросов и спама, что напрямую влияет на время реакции и качество сервиса.

Ваша команда выступает в роли разработчиков, которым поставлена задача автоматизировать этот процесс. Необходимо разработать систему, которая читает входящие обращения из файловой системы, классифицирует их по категориям и формирует отчёт о проделанной работе.

Команда получает готовую структуру директорий с набором из ~100 входящих писем. Почтовый ящик устроен так же, как в классических почтовых клиентах: есть папка входящих, а задача команды — разложить письма по смысловым папкам. Стандартные категории, которые стоит рассмотреть: входящие, отправленные, спам, важные, черновики — однако итоговый набор категорий команда определяет самостоятельно, исходя из анализа содержимого писем и соображений практической пользы.

Обратите внимание: разнообразие форматов и непредсказуемость контента во входящих письмах не случайность, а предусмотренная особенность задачи.

### Задание

Разработать систему обработки почты, включающую:

- Python-приложение с основной логикой: чтение и обработку писем, классификация по выбранным категориям, перемещение файлов в соответствующие папки, формирование лога и статистики.
- Bash-скрипт, запускающий приложение и выводящий необходимую информацию о результате работы: вывод итогового статуса обработки, базовые проверки.
- Тесты, покрывающие ключевую логику: чтение письма, классификация, обработка граничных случаев.
- Git-репозиторий, организующий совместную работу с историей коммитов, ветками и осмысленными сообщениями.

**Важно об устойчивости классификации:** одна из ключевых аналитических задач хакатона — продумать классификацию так, чтобы она была устойчивой: что произойдёт, если новое письмо не попадёт ни в одну из введённых категорий? Как система будет справляться с появлением новых типов обращений? Ответы на эти вопросы должны быть отражены как в решении, так и в презентации.

### Требования к проекту

В качестве результата команда предоставляет ссылку на GitHub-репозиторий и защищает проект перед комиссией.

Репозиторий должен содержать:

- исходный код приложения и тесты
- bash-скрипт запуска
- файл `README.md` с описанием проекта и инструкцией по запуску
- файл `.gitignore`

Формат защиты: презентация решения командой в произвольном формате, включая логику принятых решений и архитектуру. Демонстрация работающего кода — на усмотрение команды.

Оформление презентации остаётся полностью на ваше усмотрение, но помните, что результат должен быть релевантен для демонстрации бизнес-заказчику — комиссию, принимающую вашу работу, правильнее всего воспринимать именно в таком качестве. Например, вставлять в презентацию строчки кода или злоупотреблять скринами среды разработки не рекомендуется.

## Базовые критерии оценки проекта (8 баллов)

Оценка за групповой проект в целом — то есть первая из двух оценок — выставляется по 10 балльной шкале, исходя из базовых 6 критериев (критерии на оценку 9 и 10 описаны отдельно). Также комиссия на своё усмотрение может добавить (но не снизить) некоторое количество бонусных баллов за рамками данных критериев.


### Критерий 1. Анализ задачи и архитектура решения (до 2 баллов)

Данный критерий оценивает, насколько команда осмыслила поставленную задачу: как обосновала систему категорий, как спроектировала решение и как презентовала логику своей работы.

*По данному критерию можно получить **1 балл** при условии, что команда определила набор категорий и может объяснить логику их выбора, а в презентации понятно, как устроено решение и как данные проходят через систему.*

*По данному критерию можно получить **максимальный балл в 2 балла** при дополнительном выполнении следующих условий: команда продумала, что происходит в нестандартных ситуациях — например, если новое письмо не подходит ни под одну из введённых категорий или если появляется совершенно новый тип обращений; в презентации представлена схема того, как работает система (пайплайн), и сформулированы содержательные выводы по результатам обработки набора писем.*

### Критерий 2. Python-код и логика классификации (до 2 баллов)

Данный критерий оценивает корректность и качество реализации основной логики приложения.

*По данному критерию можно получить **1 балл** при условии базовой работающей реализации: письма читаются, классифицируются и перемещаются в нужные папки; используется ООП — как минимум выделены классы с разумным распределением ответственности.*

*По данному критерию можно получить **максимальный балл в 2 балла** при дополнительном выполнении следующего условия: в проекте используются инструменты или возможности Python, которые не разбирались на занятиях, — например, для более гибкой работы с данными, расширенного анализа текста или ведения логов, — и команда может объяснить, зачем выбрала именно их; код читаем и логично организован.*

### Критерий 3. Работа с файловой системой и исключениями (до 1 балла)

Данный критерий оценивает, насколько корректно приложение работает с реальными данными, которые могут содержать неожиданные форматы и проблемы.

*По данному критерию можно получить **0,5 балла** при условии, что приложение не падает на нестандартных входных данных: файлах неожиданного формата, нечитаемых или повреждённых письмах.*

*По данному критерию можно получить **максимальный балл в 1 балл** при дополнительном выполнении следующего условия: команда заранее продумала, какие проблемные ситуации вообще могут возникнуть, и для каждой из них реализовала понятное решение — например, что происходит с файлом, который не удалось прочитать, или с письмом в неизвестном формате: куда оно попадает, что записывается в лог и почему именно так.*

### Критерий 4. Bash и взаимодействие с ОС (до 0,5 баллов)

Данный критерий оценивает bash-скрипт запуска и понимание взаимодействия приложения с операционной системой.

*По данному критерию можно получить **максимальный балл в 0,5 балла** при условии, что команда написала bash-скрипт, который корректно запускает приложение и использует хотя бы одну конструкцию помимо вызова Python — например, вывод статуса, перенаправление вывода в лог-файл или проверку существования нужной директории.*

### Критерий 5. Тестирование (до 1,5 баллов)

Данный критерий оценивает наличие, качество и разнообразие тестов.

*По данному критерию можно получить **0,5 балла** при условии, что тесты покрывают ключевую логику работающего приложения, успешно проходят и используют собственные тестовые данные — то есть не зависят от конкретных файлов в папке `inbox/.`*

*По данному критерию можно получить **1 балл** при дополнительном выполнении следующего условия: проверяются граничные случаи — например, письмо без совпадений с категориями, пустой файл, файл неизвестного формата; по набору тестов видно, что команда думала о том, что может пойти не так.*

*По данному критерию можно получить **максимальный балл в 1,5 балла** при дополнительном выполнении следующего условия: тесты параметризованы или используют фикстуры; покрытие логики заметно шире минимального — видно, что тестирование воспринималось как полноценная часть разработки, а не формальное требование.*

### Критерий 6. Git и совместная работа (до 1 балла)

Данный критерий оценивает историю репозитория и культуру командной разработки.

*По данному критерию можно получить **0,5 балла** при выполнении всех следующих требований к репозиторию: наличие `README.md` с описанием проекта и инструкцией по запуску — после прочтения у стороннего человека не должно остаться вопросов о том, что делает проект и как его запустить; наличие `.gitignore`; осмысленная структура директорий.*

*По данному критерию можно получить **максимальный балл в 1 балл** при дополнительном выполнении следующего условия: по истории коммитов видно, что команда работала вместе на протяжении всего хакатона — есть вклад от нескольких участников, коммиты делались регулярно с понятными сообщениями, а не всё залито одним коммитом в последний момент; использовались ветки для разработки отдельных частей проекта.*

## Критерий 7. Расширение проекта (оценка 9–10 баллов)

Данный критерий применяется только при условии, что **критерии 1 и 2 выполнены на максимальный балл**, то есть базовая часть проекта реализована полностью и без существенных недочётов. Если это условие не выполнено, **максимальная оценка за проект составляет 8 баллов** вне зависимости от наличия расширения.

Оценки 9 и 10 баллов выставляются командам, которые помимо базовой части реализовали осмысленное расширение — что-то полезное поверх уже работающей системы. Расширение должно опираться на результаты основной логики: классификацию писем, статистику, логи или структуру выходных папок. Отдельный модуль, никак не связанный с основной системой, расширением не считается.

*Расширение может быть любым — команда сама решает, что добавить. Например, это может быть удобный отчёт по результатам обработки, визуализация статистики, возможность настраивать категории без изменения кода, режим просмотра результатов перед их применением, ML-модель как альтернативный способ классификации поверх базовой логики или что-то другое, что команда считает полезным. Это лишь примеры — приветствуется любая идея, которую команда может обосновать и объяснить.*

*Граница между базовой частью и расширением определяется просто: классификация на основе собственных правил, то есть логика, которую команда продумала и реализовала самостоятельно, является **обязательной частью** проекта и должна работать **независимо от любых расширений**. Всё что строится поверх неё: умный классификатор на основе ML-модели, интерфейс, отчёт или что-то другое — относится к расширению и может быть реализовано с помощью любых инструментов, включая генеративный ИИ. При этом команда должна уметь объяснить на защите, как работает добавленный код, зачем он нужен и почему для этой части было принято решение использовать ИИ-инструменты.*

*Использование генеративного ИИ допускается **исключительно для реализации расширения**. Применение ИИ для написания базовой части проекта: чтения писем, классификации по правилам, обработки ошибок, тестов, bash-скрипта — считается нарушением и влечёт **0 баллов** за весь проект для всех участников команды.*

*Итоговую оценку 9 или 10 баллов комиссия выставляет на своё усмотрение, исходя из того, насколько расширение продуманно, полезно и органично вписывается в проект.*
